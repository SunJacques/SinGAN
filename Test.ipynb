```python
from PIL import Image
from matplotlib import pyplot as plt
import torch
from torchvision import transforms as T
```

```python
scale_factor = 0.9
real = Image.open("img.jpeg")
real = real.convert("RGB")
real = T.ToTensor()(real).unsqueeze(0)
# print(real)
w,h = real.shape[3],real.shape[2]
real = T.Resize((int(h*scale_factor),int(w*scale_factor)))(real)
print(real.shape)
real = real.squeeze(0).permute(1,2,0).numpy()
print(real)
plt.imshow(real)
```
